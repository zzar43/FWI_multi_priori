```julia
using PyPlot
using JLD2
```

```julia
@load "Example1/lbfgs_CFP_temp_box_19.jld2"
res3_19 = reshape(u1,101,101);

@load "Example1/lbfgs_CFP_temp_box_20.jld2"
res3_20 = reshape(u1,101,101);

res3 = res3_20 - res3_19;
```

```julia
figure(figsize=(10,4))
subplot(131)
imshow(res3_19); clim(1,1.2); colorbar(shrink=0.5)
title("(a) 19 iterations")

subplot(132)
imshow(res3_20); clim(1,1.2); colorbar(shrink=0.5)
title("(b) 20 iterations")

subplot(133)
imshow(res3); colorbar(shrink=0.5)
title("(c) difference between (a) and (b)")

tight_layout()
# savefig("fig1.png")
```

```julia
println(sqrt(sum(res3_20 .* res3_20)))
println(sqrt(sum(res3 .* res3)))
```

```julia
println(sqrt(sum(res3 .* res3)) / sqrt(sum(res3_20 .* res3_20)))
println(sqrt(sum(res3 .* res3)) / sqrt(sum(res3_20 .* res3_20)) * 100)
```

```julia
@load "Example3/tv_1200/lbfgs_CFP_temp_box_tv_49.jld2"
res1200_49 = reshape(u1,76,251);

@load "Example3/tv_1200/lbfgs_CFP_temp_box_tv_50.jld2"
res1200_50 = reshape(u1,76,251);

res1200 = res1200_50 - res1200_49;
```
